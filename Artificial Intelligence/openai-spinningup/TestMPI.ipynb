```
if os.getenv("IN_MPI") is None:
        env = os.environ.copy()
        env.update(
            MKL_NUM_THREADS="1",
            OMP_NUM_THREADS="1",
            IN_MPI="1"
        )
        args = ["mpirun", "-np", str(2)]
        
        args += [sys.executable] + sys.argv
        subprocess.check_call(args, env=env)
        sys.exit()
```
